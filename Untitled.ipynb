```python
import numpy as np
import h5py

def brain_mri_mask_kykz(
    Ny,
    Nz,
    R,
    acs_size=32,
    shear_factor=1
):
    """
    Generate ky-kz subsampling mask for brain MRI

    Returns:
        mask: shape (Ny, Nz), binary
    """

    mask = np.zeros((Ny, Nz), dtype=np.float32)

    # -------------------------
    # 1. ACS region (32 Ã— 32)
    # -------------------------
    ky_center = Ny // 2
    kz_center = Nz // 2

    ky0 = ky_center - acs_size // 2
    ky1 = ky0 + acs_size
    kz0 = kz_center - acs_size // 2
    kz1 = kz0 + acs_size

    mask[ky0:ky1, kz0:kz1] = 1

    # -------------------------
    # 2. Subsampling outside ACS
    # -------------------------
    for kz in range(Nz):
        ky_offset = (shear_factor * kz) % R
        ky_lines = np.arange(ky_offset, Ny, R)

        for ky in ky_lines:
            if not (ky0 <= ky < ky1 and kz0 <= kz < kz1):
                mask[ky, kz] = 1

    return mask
```

```python
import numpy as np

def generate_kspace_mask(
    num_pe_lines,
    num_ro_lines,
    R,
    acs_lines=32
):
    """
    Generate Cartesian undersampling mask with ACS.

    Args:
        num_pe_lines (int): number of phase-encoding lines
        num_ro_lines (int): number of readout points
        R (int): acceleration factor
        acs_lines (int): number of fully sampled ACS lines

    Returns:
        mask (np.ndarray): shape (num_pe_lines, num_ro_lines)
    """

    mask_1d = np.zeros(num_pe_lines)

    # Center indices
    center = num_pe_lines // 2
    acs_start = center - acs_lines // 2
    acs_end = acs_start + acs_lines

    # Regular undersampling
    mask_1d[::R] = 1

    # Force ACS region to be fully sampled
    mask_1d[acs_start:acs_end] = 1

    # Expand to 2D (readout direction fully sampled)
    mask = np.tile(mask_1d[:, None], (1, num_ro_lines))

    return mask
```

```python
import scipy.io as sio

sio.savemat("kspace_mask.mat", {"mask": mask})
```
