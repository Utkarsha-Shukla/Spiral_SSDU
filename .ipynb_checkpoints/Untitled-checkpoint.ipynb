```python
import os
import scipy.io as sio
import numpy as np
import parser_ops

mask_dir='/home/hpc/mfdp/mfdp120h/SSDU/SSDU-main/SSDU-main/data/'

parser = parser_ops.get_parser()
args = parser.parse_args()

print(mask_dir)
if os.path.exists(mask_dir):
    print("Creating original sampling mask...")

    nrow = args.nrow_GLOB
    ncol = args.ncol_GLOB
    acc = args.acc_rate  # e.g. 4, 8

    mask = np.zeros((nrow, ncol), dtype=np.float32)

    # ---- Cartesian undersampling example ----
    center_fraction = 0.08
    num_center = int(ncol * center_fraction)
    start = (ncol - num_center) // 2
    mask[:, start:start + num_center] = 1

    # Randomly sample remaining columns
    prob = (ncol / acc - num_center) / (ncol - num_center)
    for c in range(ncol):
        if mask[0, c] == 0 and np.random.rand() < prob:
            mask[:, c] = 1

    sio.savemat(mask_dir, {'mask': mask})
    print("Saved original mask to:", mask_dir)
```
